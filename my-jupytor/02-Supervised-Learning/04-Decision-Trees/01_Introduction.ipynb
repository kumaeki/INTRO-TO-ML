# Decision Trees

At a high level, Decision Trees are easy to understand because it is very similar to how our brains work in some situations. 

The more complex a situation comes, the deeper and wider the tree becomes. 

This lesson will show you how machine learning builds and uses these to make effective predictions.



As with any real life data, not all Decision Trees are not usually completely balanced.
