# Solution : Other Activation Fuctions

### Hyperbolic tangent function

$$tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi,100)
y = np.tanh(x)

plt.plot(x, y)
plt.show()
```

### Rectified linear unit (ReLU)

```python
def relu(x):
  return x * (x > 0.0)

x = np.linspace(-np.pi, np.pi,100)
y = relu(x)

plt.plot(x, y)
plt.show()
```
