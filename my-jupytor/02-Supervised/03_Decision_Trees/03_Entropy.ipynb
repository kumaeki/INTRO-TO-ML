# The Formula for entropy when tha balls can be of two colors


When we have 5 red balls and 3 blue balls



The general Formula is :
$$ Entropy = -\frac{m}{m+n}{log}_2(\frac{m}{m+n})-\frac{n}{m+n}{log}_2(\frac{n}{m+n}) $$

let's state the probabilities $p1$ and $p2$ , instead of the number of balls

$$ p_1 = \frac{m}{m+n}$$

$$ p_2 = \frac{n}{m+n}$$

We got

$$ Entropy = -p_1{log}_2(p_1)-p_2{log}_2(p_2) $$

# Multi-class Entropy

$$ Entropy = -p_1{log}_2(p_1)-p_2{log}_2(p_2)-...-p_n{log}_2(p_n) = -\sum_{i=1}^{n}p_i{log}_2(p_i) $$
