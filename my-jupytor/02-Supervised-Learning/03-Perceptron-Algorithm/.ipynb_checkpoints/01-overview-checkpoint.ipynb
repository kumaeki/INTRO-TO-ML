# Classification Algorithms

Classification Algorithm are those that answer questions that have answers, yes or no.

For example, is ths email spam or not?

The First one of these algorithms is the Perceptron Algorithm

At a high level, this lesson focus on the idea of a perceptron, is a node that accepts various input and ultimately produces a signle output.

The inputs can represent the many dimensions of the dataset.

All of these inputs are them plotted as a point in a graph to see whether the points is plotted in the positive area.

If so, the output is YES, otherwise is NO.

# Acceptacne at a University

Imagine we are the admissions office at a university and our job is to accept or reject students. 

So, in order to evaluate students, we have two pieces of information,

- The results of a test
- Their grades in school

First, we label the horizontal axis corresponding to the test by the variable $x_1$x , and the vertical axis corresponding to the grades by the variable $x_2$.

 So this boundary line that separates the blue and the red points is going to have a linear equation. The one drawn has equation 
$$ 2x_1+x_2-18=0 $$

means 

$$score = 2∗Test+Grades−18$$

This is called a prediction.

- Score > 0: Accept
- Score < 0: Reject

We'll abbreviate this as $Wx+b=0$, where $W =(w_1,w_2)$ and $x = (x_1,x_2)$ and we take the product of the two vectors.

- $x$ is the input
- $W$ is weight
- $b$ is the bias


# Higher Dimensions

What if we have many columns? 

Our data just leaps in n-dimensional space. 

If we can imagine that the points are just things with n coordinates called $x_1$, $x_2$, $x_3$, up to $x_n$

with our $\hat{y}$, then our boundary is just an $(n-1)$-dimensional hyperplane, 

which is a high dimensional equivalent of a line in 2D or a plane in 3D. 

And the equation of this $(n-1)$-dimensional hyperplane is going to be 
$$w_1x_1 +w_2x_2 +...+ w_nx_n +b =0$$

which we can still abbreviate as 

$$Wx+b=0$$

where our vectors now have n entries.

Our prediction is still the same as before. It is

- $\hat{y}=1$ if $Wx + b\geq0$

- $\hat{y}=0$ if$ Wx + b\lt0$
