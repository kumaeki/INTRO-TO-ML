# The probability of sick

Imagine we're not feeling very well and we go to the doctor

The doctor says there's a terrible disease going on, and she'll administer a test for you. 

Moreover, she says that the test has 99 percent accuracy.

- For every 100 patients that are sick, the test correctly diagnosis 99 of them,

- For every 100 patients that are healthy, the test correctly diagnosis 99 of them.

We researched the Internet and find that on average, one out of every 10,000 people suffers from the disease.

So, When doctor says that we have tested positive for the disease, what is the probability of we are sick?
