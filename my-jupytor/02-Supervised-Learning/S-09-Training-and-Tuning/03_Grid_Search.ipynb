# Machine learning summary

1. We train a bunch of models with our training data
1. We use cross-validation data to pick the best of these models
1. We test it with the testing data to make sure our model is good.

### Logistic regression model example
Let's say we have four candidates. We train a model of degree one, a model of degrees two, three, and four.

1. We train them with the training data to find the slope and the coefficients of the polynomials et cetera.
1. We use the cross-validation data to calculate say, the F1 score of all these models.
1. We pick the model with the highest F1 score.
1. As a final step, we use our testing data to make sure our model is good.

The parameters of the algorithm in this case are the coefficients of the polynomial but the degree of the polynomial is called a **hyperparameter**.



### Decision tree model example

What are the hyperparameters? Well, one of them is the depth with values of one, two, three, and four.

1. We use the training data to train a bunch of trees of depth one, two, three, and four.
1. The parameters here are the thresholds in the leaves and the nodes et cetera.
1. Then we take the F1 score and calculate it on the cross-validation set on each of these models.
1. Then we pick the one that did the best
1. Finally, with the testing set, we make sure the model is good.



### Support vector machine model example
In an SVM, we have some hyperparameters like the kernel which can be linear or polynomial, and we also have the gamma parameter.

#### Grid search
Make a table with all the possibilities and pick the best one. Our columns here are the different kernels we can use; linear and polynomial. Our rows are the different values of gamma. It's recommended to take a few values that grow exponentially such as 0.1, 1, 10,100,1000 et cetera.

1. We use a training set to train a bunch of linear models and polynomial models with different values of gamma
1. We use the cross-validation set to calculate the F1 score in all of these models and then we simply pick the one with the highest F1 score.
1. We use the testing set to make sure that what we did was good.
