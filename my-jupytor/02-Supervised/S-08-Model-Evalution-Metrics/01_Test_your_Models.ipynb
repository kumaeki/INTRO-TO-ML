## when Accuracy don't work

**Precision** focuses on the predicted "positive" values in your dataset.






| |positive|negative|
|----|----|----|
|positive |True Positive(TP)|False Positive(FP)|
| Negative|False Negative(FN)|True Negative|



$$Precision = \frac{(True Positive)}{(TP+FP)}$$

**Recall** focuses on the actual "positive" values in your dataset



| |positive|negative|
|----|----|----|
|positive |True Positive(TP)|False Positive(FP)|
| positive|False Negative(FN)|True Negative|



$$Precision = \frac{(True Positive)}{(TP+FN)}$$

## combine this two numbers into one

**$F_1$ score(Harmonic mean)**

$$F_1 = 2\frac{xy}{x+y}$$

**$F_\beta$ score**

The $\beta$ parameter controls the degree to which precision is weighed into the F score, which allows precision and recall to be considered simultaneously. 


$$F_\beta = (1 + \beta^2) \times \frac{Precision \times Recall}{(\beta^2 \times Precesion) + Recall}$$

**Boundaries in the $F_\beta$ score**

- If $\beta= 0$ , then we get precision. $$F_0 = (1 + 0^2)\times \frac{Precision \times Recall}{(0^2 \times Precesion) + Recall} = Precision$$

- If $\beta = \infty$, then we get recall.$$F_0 = \lim\limits_{\beta\rightarrow\infty}((1 + \beta^2) \times \frac{Precision \times Recall}{(\beta^2 \times Precesion) + Recall} = \lim\limits_{\beta\rightarrow\infty}(\frac{Precision \times Recall}{(\frac{\beta^2}{1 + \beta^2} \times Precesion) + \frac{1}{1 + \beta^2} \times Recall} = \frac{Precision \times Recall}{(1\times Precesion) + 0 \times Recall} = Recall$$

- If $\beta = 1$ then we get the harmonic mean of precision and recall.

- For other values of $\beta$, if they are close to 0, we get something close to precision, if they are large numbers, then we get something close to recall
