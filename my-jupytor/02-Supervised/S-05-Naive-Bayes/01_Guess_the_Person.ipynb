# The Problem

There is an office with two people, Alex and Brenda, and they're both there the same amount of time.

When they were in the office and we see someone passing by really fast, we can't tell who it is, but we'd like to take a guess.

**Both of the probability are 50%**

Then, we have some new information.

- Alex wears a red sweater two days a week
- Brenda wears a red sweater three days a week

If we saw a person pass by five times

**The probability of Alex is 40%, and the probability of Brenda is 60%**



## This is the essence of Bayes' theorem.

- The initial guess we had, the 50/50 guess, is called the prior since it's all we could infer prior to the new information about the red sweater.

- The final guess we have, the 60/40 guess is called the posterior since we've inferred it after the new information has arrived.

## What we know vs. what we infer

We know the probability of an event A, and some more information, a new event R which releated to A (The probability of R given A).

The Bayes's theorem infers the probability of A given R



# guess the person again

We know:

- Alex wears a red sweater two days a week
- Brenda wears a red sweater three days a week

Now we get more information about Alex and Brenda

- Alex comes to the office three days a week.
- Brenda comes to the office a days a week.

How do we calculate the probability?



# The formal version of the Bayes Theorem





### when we know the probability of event A and B, and third event R, we get the probability of  A given R

$$ P(A|R) = \frac{P(A)*P(R|A)}{P(A)*P(R|A) + P(B)*P(R|B)} $$
