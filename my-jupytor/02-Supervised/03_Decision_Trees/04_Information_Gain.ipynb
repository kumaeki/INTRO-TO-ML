# Information Gain

we have there way to split the red one and bule one away.

How do we evaluate which one is the best?



Use **Information Gain**

The Information Gain is the change in entropy



$$InformationGain = Entropy(Parent)[ \frac{m}{m+n}Entropy(Child1) +  \frac{n}{m+n}Entropy(Child 2)]$$

The last one have the biggest information gain, so it's the best one!
