# Hyperparameters for Decision Trees


In order to create decision trees that will generalize to new problems well, we can tune a number of different about the trees.

We call them **Hyperparameters**

 ### Maximum Depth
 
 The Maximum Depth of a decision tree is simply the largest possible length between the root to a leaf.
 
 A tree of maximum length $k$ can have at most $2^k$ leaves.



### Minimum number of samples to split

A node must have at least <kbd>min_samples_split</kbd> samples in order to be large enough to split.

If the samples are fewer than <kbd>min_samples_split</kbd> , it will not be split, and the process stops.

### Minimum number of smples per leaf

The number <kbd>min_samples_leaf</kbd> can be specified as an **integer** or a **float**.

If it's a integer, it's the minimum number of samples allowed in a leaf.

If it's a float, it's  the minimum percentage of samples allowed in a leaf.
