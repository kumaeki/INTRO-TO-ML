# Naive assumptions

In naive Bayes, we assume that our probabilities are independent.

This is a false and naive assumption, but in practise, it works very well and it makes our algorithm very fase

# easy moneny

Assume we have an email that contains the words 'easy' and 'money'. 

We want to determine the probability of the email being spam given that it contains these words.

$$ P(spam  easy  ,  money  )P(  easy  ,  money  spam)P(spam)  =P(  easy spam) P( money  spam)P(spam)  $$



$$ P(spam'easy','money')=\frac{1}{12}$$

$$ P(ham'easy','money')=\frac{1}{40}$$

# In general

Let's say we have N suspicious words. 

For example, easy, money, cheap, et cetera.

These are our conditions, and we get a new email we tested against them. 

we make the Naive assumption that the probability of spam given that the email contains all these words is the product of these probabilities
