# Perceptron Trick

Let's say we have our linear equation, for example, $ 3x_1 + 4x_2 -10$.

And that linear equation gives us a line which is the points where the equation is zero and two regions. 

The positive region drawn in blue where $ 3x_1 + 4x_2 - 10$ is positive, and the negative region drawn in red with $ 3x_1 + 4x_2 - 10$  is negative.

Assume we have a misclassified point $ (4,5) $ which is a red point in the blue area.

So how do we get that point to come closer to the line?



This line will move drastically towards the point, possibly even going over it and placing it in the correct area
